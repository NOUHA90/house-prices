```python
# Import library
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt # Visualization
import seaborn as sns #Visualization
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn import set_config
set_config(display='diagram')
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
house_prices_train = pd.read_csv('house_prices_cleaned_train_data.csv')
house_prices_test = pd.read_csv('house_prices_cleaned_test_data.csv')
train_ID = house_prices_train.Id
test_ID = house_prices_test.Id
target = pd.read_csv('target.csv')
house_prices_train.shape, house_prices_test.shape, target.shape
```

```python
house_prices_train = house_prices_train.drop(columns=['Unnamed: 0'])
```

```python
house_prices_test = house_prices_test.drop(columns=['Unnamed: 0'])
```

```python
target = target.drop(columns=['Unnamed: 0'])
```

```python
house_prices_train.shape, house_prices_test.shape, target.shape
```

```python
house_prices_train.head()
```

```python
house_prices_test.head()
```

```python
numeric_features = ['OverallQual', 'GrLivArea', 'GarageArea', 'TotalBsmtSF']
numeric_transformer = Pipeline([('impute', SimpleImputer(strategy='mean'))])
numeric_transformer
```

```python
preprocess_pipeline = ColumnTransformer([('numeric_features', numeric_transformer, numeric_features)])
preprocess_pipeline
```

```python
test = preprocess_pipeline.fit_transform(house_prices_test)
```

```python
test = pd.DataFrame(test, columns = ['OverallQual', 'GrLivArea', 'GarageArea', 'TotalBsmtSF'])
test
```

```python
linreg = LinearRegression()
linreg.fit(house_prices_train[['OverallQual', 'GrLivArea', 'GarageArea', 'TotalBsmtSF']], target)
```

```python
 # model evaluation for testing set
y_train_predict = linreg.predict(house_prices_train[['OverallQual', 'GrLivArea', 'GarageArea', 'TotalBsmtSF']])
print(y_train_predict)
```

```python
 # model evaluation for testing set
y_test_predict = linreg.predict(test)
print(y_test_predict)
```

```python
y_test_predict.shape
```

```python
y_test_predict = pd.DataFrame(y_test_predict, columns = ["SalePrice"])
```
